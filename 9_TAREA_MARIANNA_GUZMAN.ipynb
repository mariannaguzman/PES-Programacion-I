## Ejercicio 1

1. Generar arreglos de numpy para obtener el resultado de $ z = f(x,y) $ para un conjunto de entrada aleatorio de al menos 100 filas.

- Crear un `DataFrame` de pandas con las columnas $x$, $y$ y $z$.
- Obtener el promedio de cada `Series` en el `DataFrame`.

```python
# Tu código acá
#INCISO 1
matriz=np.random.randint(10,size=[100,2])#Genero una matriz con números aleatorios, 100 filas, 2 colum
#2 columnas por ser x, y
print(matriz)
def f(matriz):
    return (matriz[::,0])**0.6*(matriz[::,1])**0.4 #Defino una función en donde la posición 0=x y posición1=y
```

```python
z = pd.Series(f(matriz)) #Mi variable z es una serie de la función sobre la matriz que meta.
z
```

```python
#INCISO 2
x=pd.Series(matriz[::,0]) #Genero una serie para mi variable x
y=pd.Series(matriz[::,1]) #Genero una serie para mi variable y
```

```python
df = pd.DataFrame({ 'X': x, 'Y': y, 'Z':z }) #Le doy un dataframe que me saque x,y,z
df
```

```python
#INCISO 3
df.mean() #Saco la media a x,y,z
```

```python
import matplotlib
import matplotlib.pyplot as plt
#Array
x1=np.random.randint(10,size=(100,1)) #Obtengo un array de números aleatorios de 1 columna:variable X

def f2(vector):
    return vector*2 #Defino mi función

y1= f2(x1)#Obtengo un array que es la función aplicada a la X
```

```python
#FORMO MI GRÁFICA

plt.figure(figsize=(8, 6)) #tamaño

plt.plot(x1, y1, '#f39200')#plotear x, y

plt.xlabel('x') #Para onerle nombre a cada eje
plt.ylabel('y')

plt.title(r"Plot of $y=2$x") #para ponerle nombre a la gráfica
plt.show() #que muestre la gráfica
```

```python
#DATA FRAME
num=np.random.randint(10,size=(100,1)) #Me da una columna de num aleatorios
x2=pd.Series(num.flatten()) #Le doy flatten para estar en 1 dimensión y le doy series.
x2
```

```python
def f(vector):
    return vector*2 #Defino una función
y2=f(x2)#Le aplico la función a y y me devuelve una serie.
y2
```

```python
dataframe=pd.DataFrame({ 'X': x2, 'Y': y2 })
dataframe #Creo un data frame con mi nueva x,y
```

- Haga una gráfica de barras para visualizar la tarifa promedio pagada por los pasajeros de acuerdo a su edad. Para esto:
    - Primero, separamos a los pasajeros por rango, utilizando la función `pd.cut` y agregamos esta serie al DataFrame.
    - Debemos agrupar por esta columna y calcular el promedio de las tarifas.
    - Por último, agregamos `plot(kind='bar')` para obtener la gráfica de barras.

```python
df['AgeClass'] = pd.cut(df['Age'], bins=np.arange(0,90,10)) #Me agregó una columna del rango de edades
df
# Tu código acá
```

```python
age_class=df.groupby("AgeClass")["Fare"].mean() #Le saco la media de las tarifas por rango de edaades
age_class
```
